```python
import pandas as pd
```

```python
def load_data():

    df = pd.read_csv('../data/creditcard.csv')
    df = df.drop_duplicates().reset_index(drop=True)

    df.columns = df.columns.str.lower()
    df = df.rename(columns={"class": "fraud"})
    
    del df['time']

    print('Data was loaded successfully')
    return df
df = load_data()
```
